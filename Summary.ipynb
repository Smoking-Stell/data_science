# Основы

Центральная предельная теорема

\begin{equation}
\frac{\sum_{i=1}^{n} X_i - n\mu}{\sqrt{n}\sigma} \xrightarrow{n \to \infty} N(0, 1)
\end{equation}


Стандартное отклонение распределния выборочных средних (ошибка среднего)
\begin{equation}
se = \frac{σ}{\sqrt{n}}
\end{equation}


Доверительные интервалы

\begin{equation}
\bar{X} \pm z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}}
\end{equation}


z-оценка

\begin{equation}
z = \frac{X - \mu}{se}
\end{equation}

оно выдает то насколько стандарнтных отклонений тестируемое значение было бы вправо или вево от оригинального

t-распредление

\begin{equation}
z = \frac{X - \mu}{\frac{sd}{\sqrt{n}}}
\end{equation}

его применяем когда выборка маленькая или когда не знаем оригинальное стандартное отклонение


P уровень значимости обозначает с какой вероятностью наше или более сильное отклонение существовало. То есть если p val меньше порога то мы отклоняем нулевую гипотезу о равенстве

\begin{equation}
t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{sd_1^2}{n_1} + \frac{sd_2^2}{n_2}}}
\end{equation}

теперь тут учитывается что их два

SSB -внешне групповая изменчивость (between)
SSW - внутри групп (within)

\begin{equation}
F = \frac{ \frac{SSB}{k-1} }{ \frac{SSW}{n-k} }
\end{equation}

n - количество групп, k - количество наблюдений (в формуле это все степени свободы бтв)

ошибка первого рода - H1 неверно принята

ошибка второго рода - H0 неверно принята


Формула корреляции

\begin{equation}
\rho_{xy} = \frac{\text{cov}(\text{rank}(x), \text{rank}(y))}{\sigma_{\text{rank}(x)} \sigma_{\text{rank}(y)}}
\end{equation}

Фрмула для коэфициента ковариации

\begin{equation}
r_{xy} = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2 \sum_{i=1}^{n} (y_i - \bar{y})^2}}
\end{equation}

коэф детерминации - \(r^2) показывает какая часть итоговой переменной зависит от параметра


формулы коэфициентов линейной регрессии

\begin{equation}
\beta_1 = r \cdot \frac{\text{sd}(y)}{\text{sd}(x)}
\end{equation}

\begin{equation}
\beta_0 = \bar{y} - \beta_1 \bar{x}
\end{equation}

Кси квадрат это по сути расстояния до точек в пространстве но так-то это сумма k стандартных номральных величин

\begin{equation}
\chi^2 = \sum_{i=1}^{r} \sum_{j=1}^{c} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
\end{equation}


Распределние горка от нуля и дальше убывает. При росте числа степеней свободы распределение становится более симетричным, стремится к нормальному.

Мы считаем степени свободы потому что еще одно измерение как правило полностью зависит от остальных (решка от орла, пятерка от 1,2,3,4,6 на кубике). Хотя и считаем в сумму кси квадрата отличие всех измерений

Если перемножить количество в одной графе на голичество в другой и поделить на общее число измерений, то мы полуичм нужную пропорцию

Например, если у нас 16 парней и 20 боксеров а всего человек 31, то нам нужно 16*20/31 и так получим нужную пропорцию в идеальном мире

Нужно применять поправку Йетса и вычитать 0.5 из расстояния до возведения в квадрат

Также стоит учитывать, что степени свободы считаются по формуле (n-1)(m-1)





```python
from scipy import stats
tab = ([20,15],[11,12],[7,9])

chi2, p_value, dof, ev = stats.chi2_contingency(tab)
print(f'''
Хи квадрат {chi2}
p - value {p_value}
Степеней свобод {dof}
Ожидаемые наблюдения
{ev}
''')
```

Есть специальная таблица сопряженности в табоицах, которая показывает отклонения, ноглобально это просто визуализация

Критерий фишера применяется только на совсем маленьких выборках. Мы рассчитываем вероятность получения такой или еще более сильновыраженных изменений. По сути просто комбинаторика

Логистическая регрессия. Зависимая переменная - две градации строго. Две градации для уравнивания частей уравнения мы переводим в логарифм шансов (ln(поз. исхода/нег. исходы))

Intercept  = коэфициент который по считается через логарифм шансов и показывает вероятность без предикторов

стандартные отклонения и z-value как обычно показывает размер стандартного отклонения и сколько их до нашего числа

При добавлении предикторов мы считаем отношение шансов одной группы к другой, которая начальная-intercept и так получаем коэфициент (если предиктор качественный). При добавлении новых предикторов мы получаем новую размерность которая теперь тоже считается как отношение к интерсепту.

Как только добавляем пересечение предикторов не входящих в интерсепт нужно считать поправку вычитая из отношения этого пересечения отношение на первом уровне (один предиктор входит в интерсепт, другой берется такой же как в новом вводимом) (ТЯЖЕЛО, 2.5 урок со второй статистики)

При добавлении числовых предикторов (до этого напомню были номинативные), появляется доп коэф, который в интерсепте равен 0
